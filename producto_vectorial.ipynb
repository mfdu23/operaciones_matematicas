```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
v1= np.array([3, 4, 5])
v2 = np.array([1, 2, 3])

producto_escalar = np.dot(v1, v2)

producto_vectorial = np.cross(v1, v2)

norm_v1 = np.linalg.norm(v1)
norm_v2 = np.linalg.norm(v2)

cos_angulo = producto_escalar / (norm_v1 * norm_v2)
angulo = np.arccos(np.clip(cos_angulo))

angulo_grados = np.degrees(angulo)

print(producto_escalar, producto_vectorial, angulo_grados)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='Vector 1')
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='b', label='Vector 2')

ax.quiver(0, 0, 0, producto_vectorial[0], producto_vectorial[1], producto_vectorial[2], color='g', label='Producto Vectorial')

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)

plt.legend()
plt.title('Producto Escalar y Vectorial en 3D')

plt.show()
```
