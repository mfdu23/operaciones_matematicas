```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v1=np.array([3,4])
v2=np.array([1,2])

suma=v1+v2
resta=v1-v2
producto_escalar=np.dot(v1,v2)

print("Suma:", suma)
print("Resta:", resta)
print("Producto escalar:", producto_escalar)
```

```python
fig,ax=plt.subplots()
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')

ax.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1, color='g', label='Suma')
ax.quiver(0, 0, resta[0], resta[1], angles='xy', scale_units='xy', scale=1, color='m', label='Resta')

ax.set_xlim(-1, 7)
ax.set_ylim(-1, 7)
ax.axhline(0, color='white',linewidth=0.5, linestyle="-")
ax.axvline(0, color='white',linewidth=0.5, linestyle="-")
ax.grid(True, which='both')

plt.legend()
plt.title('Operaciones con Vectores')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
v1=np.array([3,4,5])
v2=np.array([1,2,3])

suma=v1+v2
resta=v1-v2
producto_escalar=np.dot(v1,v2)

print(suma,resta,producto_escalar)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='Vector 1')
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='b', label='Vector 2')

ax.quiver(0, 0, 0, suma[0], suma[1], suma[2], color='g', label='Suma')
ax.quiver(0, 0, 0, resta[0], resta[1], resta[2], color='m', label='Resta')

ax.set_xlim(-1, 7)
ax.set_ylim(-1, 7)
ax.set_zlim(-1, 7)

plt.legend()
plt.title('Operaciones con Vectores 3D')

plt.show()
```
